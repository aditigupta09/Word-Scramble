```python
import tkinter
from tkinter import *
import random
from tkinter import messagebox
import pyttsx3

answers = [
    "python",
    "java",
    "canada",
    "brazil",
    "london",
    "tokyo",
    "japan",
    "india",
    "australia",
    "swift"
    ]

words = [
    "ythpno",
    "vjaa",
    "adaacn",
    "zbliar",
    "odonln",
    "ookty",
    "njaap",
    "aniid",
    "saalrutia",
    "stifw"
]
engine = pyttsx3.init() 
num = random.randrange(0,10,1)

def res():
    global words,answers,num
    num = random.randrange(0,10,1)
    l1.config(text = words[num])
    e1.delete(0,END)

def default():
    global words,answers,num
    l1.config(text = words[num])
    
def check():
    global words,answers,num
    var = e1.get()
    if var==answers[num]:
        messagebox.showinfo("success!!!!!!","COOL! BRAVO! SMART SOUL! YOU GOT A POINT!")
        engine.say("COOL! BRAVO! SMART SOUL! YOU GOT A POINT!") 
        engine.runAndWait()
        res()
    else:
        messagebox.showinfo("error!!","oooopsieeeee! WRONG ANSWER")
        engine.say("BETTER LUCK NEXT TIME!!") 
        engine.runAndWait()
        e1.delete(0,END)
        
    
    
    
root = Tk()
root.geometry("350x350")
root.title("jumbled words")
root.configure(bg = "#AE1438")  #to set the background of the window

l2 = Label(root, text="LET'S PLAY!!!!", font=("droid",18), bg="#AE1438" ,fg="white")
l2.pack(padx=10, pady=10, ipadx=5, ipady=5)

l1 = Label(root, text="LET'S PLAY!!!!", font=("droid",18), bg="#AE1438" ,fg="white")
l1.pack(padx=30, pady=30, ipadx=5, ipady=5)


ans1 = StringVar()
e1 = Entry(root,font=("Arial",16), bg="white", textvariable = ans1)
e1.pack(padx=50, ipadx=5, ipady=5, pady=10)

btn1 = Button(root,text="CHECK", font=("Times New Roman",16), bg="#FFF222", fg="#333945", width=10, relief="sunken", command=check)
btn1.pack(padx=50, pady=10)

reset = Button(root,text="RESET", font=("Times New Roman",16), bg="#FF3E4D", fg="#333945", width=10, relief="sunken", command=res)
reset.pack(padx=50, pady=10)



default()

mainloop()
```
